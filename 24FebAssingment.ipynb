Q5. How are DataFrame.size() and DataFrame.shape() different?

DataFrame.size():
1. The DataFrame.size method returns the total number of elements in the DataFrame.
2. It calculates the size by multiplying the number of rows by the number of columns.
3. It returns a single integer value representing the total number of elements in the DataFrame.

```python
df1.size
```

DataFrame.shape:

1. The DataFrame.shape attribute returns a tuple representing the dimensions of the DataFrame.
2. It returns a tuple in the format (number_of_rows, number_of_columns).
3. It does not count the total number of elements but provides information about the structure of the DataFrame.

```python
df1.shape
```

Q7. You have a Pandas DataFrame df that contains a column named 'Email' that contains email addresses in the format 'username@domain.com'. Write a Python function that creates a new column 'Username' in df that contains only the username part of each email address.

The username is the part of the email address that appears before the '@' symbol. For example, if the email address is 'john.doe@example.com', the 'Username' column should contain 'john.doe'. Your function should extract the username from each email address and store it in the new 'Username' column.

```python
data3={"Email":["CoreyMSchafer@gmail.com", 'JaneDoe@email.com', 'JohnDoe@email.com']}
df5=pd.DataFrame(data3)
```

```python
df5
```

```python
df5["Username"]=df5["Email"].apply(lambda x: x.split("@")[0])
```

```python
df5
```

Q10. Given a Pandas DataFrame df with a column 'Sales' and a column 'Date', write a Python function to create a new column 'MovingAverage' that contains the moving average of the sales for the past 7 days for each row in the DataFrame. The moving average should be calculated using a window of size 7 and should include the current day.

```python
date=pd.date_range(start="2023-10-01", end="2023-10-30")
```

```python
date
```

```python
data6={"Sales":[x for x in range(100,(len(date)+1)*100,100)]}
```

```python
data6
```

```python
df9=pd.DataFrame(data6, index=[x for x in range(1, len(date)+1)])
```

```python
df9.head()
```

```python
df8=pd.DataFrame({"Date":date}, index=[x for x in range(1, len(date)+1)])
```

```python
df8.head()
```

```python
df10=df8.join(df9)
```

```python
df10.head()
```

```python
df10["Moving Average"]=df10["Sales"].rolling(window=7).mean()
```

```python
df10
```
