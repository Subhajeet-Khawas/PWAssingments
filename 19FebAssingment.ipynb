Q1. What is Flask Framework? What are the advantages of Flask Framework?

Flask is a micro web framework written in Python. It is classified as a microframework because it does not require particular tools or libraries. It has no database abstraction layer, form validation, or any other components where pre-existing third-party libraries provide common functions. However, Flask supports extensions that can add application features as if they were implemented in Flask itself.

Flask is a popular choice for developing web applications because it is easy to learn and use. It is also very flexible and can be used to build a wide variety of applications. Some of the advantages of using Flask include:
1. Easy to learn and use: Flask is a very simple framework to learn and use. It has a small API and a clear and concise documentation.
2. Flexible: Flask is a very flexible framework. It can be used to build a wide variety of applications, from simple websites to complex web applications.
3. Extensible: Flask supports a wide variety of extensions. These extensions can be used to add additional features to your application, such as database support, form validation, and authentication.
4. Active community: Flask has a large and active community of developers. This community provides support and resources for Flask developers.

Q3. What is App routing in Flask? Why do we use app routes?

App routing is a way to map specific URLs to functions in your Flask application. This allows you to control what happens when a user visits a particular URL. For example, you could use app routing to define a URL that displays a homepage, a URL that displays a list of products, and a URL that allows users to log in.

There are a few reasons why you might want to use app routing in your Flask application. First, it makes your application more organized and easier to maintain. Second, it allows you to reuse code. Third, it can help you improve your application's performance.

To use app routing in Flask, you use the route() decorator. The route() decorator takes two arguments: the URL pattern and the function that should be called when the URL is matched.

Q5. What function is used in Flask for URL Building? Write a Python code to demonstrate the working of the url_for() function.

In Flask, the url_for() function is used for URL building. The url_for() function generates a URL for a specified endpoint (view function) based on its name. It takes the endpoint name as an argument and returns the URL associated with that endpoint.

```python
pip install flask
```

```python
from flask import Flask, url_for

app = Flask(__name__)

@app.route('/')
def index():
    return 'Home Page'

@app.route('/about')
def about():
    return 'About Page'

@app.route('/user/<username>')
def user_profile(username):
    return f'Profile page of {username}'

@app.route('/contact')
def contact():
    return 'Contact Page'

if __name__ == '__main__':
    with app.test_request_context():
        # Generating URLs using url_for()
        home_url = url_for('index')
        about_url = url_for('about')
        user_url = url_for('user_profile', username='john')
        contact_url = url_for('contact')

        # Printing the generated URLs
        print('Home URL:', home_url)
        print('About URL:', about_url)
        print('User URL:', user_url)
        print('Contact URL:', contact_url)
```
